```python
import pandas as pd
import matplotlib.pyplot as plt
import os

income_data = r"C:\Users\USER\Documents\Internship\Expense Tracker\CSV\income.csv"
expenses_data = r"C:\Users\USER\Documents\Internship\Expense Tracker\CSV\expenses.csv"
savings_data = r"C:\Users\USER\Documents\Internship\Expense Tracker\CSV\savings.csv"
```

```python
#------------------Load All the necessary data----------------------

#Loading income data
def income_load():
    if os.path.exists(income_data):
        income_df = pd.read_csv(income_data)
        return income_df.iloc[0, 0]
    else:
        print(f"Income file '{income_data}' not found!")
        return 0

#Loading savings data
def savings_load():
    if os.path.exists(savings_data):
        savings_df = pd.read_csv(savings_data)
        return savings_df.iloc[0,0]
    else:
        print(f"Savings file '{savings_data}' not found!")
        return 0

#Loading expenses data
def expenses_load():
    if os.path.exists(expenses_data):
        expenses_df = pd.read_csv(expenses_data, names=['Category', 'Amount', 'Timestamp' ])
        return expenses_df
    else:
        print(f"Expenses file '{expenses_data}' not found!")
        return pd.DataFrame(columns=['Category', 'Amount', 'Timestamp'])
```

```python
#Creating Pie Chart for Expenses

def expenses_visualization():
    expenses_df = expenses_load()

    if not expenses_df.empty:
        plt.figure(figsize=(5, 5))  #Sets up the pie size to be 5x5 inches
        plt.pie(expenses_df['Amount'], labels=expenses_df['Category'], autopct='%1.1f%%', startangle=140)
        plt.title("Expense's Category Piechart\n")
        plt.axis('equal')
        plt.show()
    
    else:
        print("No available expenses") 
```

```python
#Utilizing BarChart to compare income & expenses

def expenses_income_visualization():
    income = income_load()
    expenses_df = expenses_load()

    if not expenses_df.empty:
        total_expenses = expenses_df['Amount'].sum()

        labels = ['Income', 'Expenses']
        values = [income, total_expenses]

        plt.figure(figsize=(6, 4))
        plt.bar(labels, values, color=['blue', 'yellow'])
        plt.title("Income VS Expenses Bar Chart\n")
        plt.ylabel('Amount')
        plt.show()
    
    else:
        print("No available data to compare")
```

```python
# Bar Chart to compare Savings and Income

def savings_income_visualization():
    income_df = pd.read_csv(income_data)
    savings = savings_load()
    income = income_load()

    if not income_df.empty:

        labels = ['Savings', 'Income']
        values = [savings, income]

        plt.figure(figsize=(6, 4))
        plt.bar(labels, values, color=['purple', 'red'])
        plt.title("Savings VS Income Bar Chart\n")
        #plt.ylabel('Total Amount')
        plt.show()

    else:
        print("No available income")
```

```python
#Bar Chart to show each expense category

def expenses_category_visualization():
    expenses_df = expenses_load()

    if not expenses_df.empty:
        plt.figure(figsize=(8,6))
        plt.bar(expenses_df['Category'], expenses_df['Amount'], color='pink')
        plt.title("Expense Category Bar Chart\n")
        plt.xlabel("Category")
        plt.ylabel("Amount")
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        plt.show()
        
    else:
        print("No available expense to show")
```
